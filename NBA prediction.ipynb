```python
import pandas as pd
from keras.models import Sequential
from keras.optimizers import SGD, Adam, Adadelta, RMSprop
from keras.layers import Dense
from keras.utils import to_categorical
import keras.backend as K
```

```python
df = pd.read_csv("nba_logreg.csv")
df.head()
```

```python
pd.value_counts(df["TARGET_5Yrs"])/df["TARGET_5Yrs"].count()
```

```python
X = df.drop(["TARGET_5Yrs","Name","3P%"],axis=1)
y = df["TARGET_5Yrs"]
X.shape
```

```python
X.isnull().any()
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
Xmms = mms.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(Xmms, y)
```

```python
model = Sequential()
model.add(Dense(18,input_shape=(18,),activation="tanh"))
model.add(Dense(72))
model.add(Dense(36))
model.add(Dense(1, activation='sigmoid'))
model.compile(Adam(lr=0.02),
              loss='binary_crossentropy',
              metrics=['accuracy'])
model.fit(X_train, y_train, epochs=2500, validation_data=(X_test,y_test))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC, LinearSVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold,cross_val_score
```

```python
models = [LogisticRegression(),SVC(),
          AdaBoostClassifier(),RandomForestClassifier(),
          GradientBoostingClassifier(),GaussianNB(),
          LinearSVC(),KNeighborsClassifier(),
          Perceptron(),SGDClassifier(),
          DecisionTreeClassifier()
            ]
names_models = ['Logistic Regression','Support Vector Machines',
          'Ada',  'Random Forest', 
           'Gradient Boosting Classifier', 'Naive Bayes',
          'Linear SVC','K-Neighbors',
          'Perceptron','Stochastic Gradient Decent',
          'Decision Tree Classifier'
         ]     
score = []

for model in models:
    model = model
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    acc = cross_val_score(model, X, y , cv=50)
    acc = acc.mean()
    score.append(acc)
result = pd.DataFrame({
    'model':names_models,
    'score':score
})
print (result.sort_values(by="score",ascending=False))
```
